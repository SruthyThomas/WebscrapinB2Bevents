```python
#import necessary libraries
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
#take the url rom which data needs to be scraped
url='https://www.zuddl.com/blog/fantastic-b2b-marketing-conferences-to-attend-in-2024#sep-b2b'
response=requests.get(url)
response
```

```python
#create soup object
soup=BeautifulSoup(response.content,'html.parser')
print(soup)
```

SCRAPING EVENT NAMES

```python
#find all the h3tags
h3tags=soup.find_all('h3')
# Extract specific <h3> tags by indexing
event_names = [tag.text.strip() for tag in h3tags[14:19]]
```

```python
#scraping event names
event_names
```

```python
#finding the ptags
ptags=soup.find_all('p')
ptags
```

SCRAPING EVENT DATES

```python
p_tags = soup.find_all('p')

# Initialize a list to store event dates
event_dates = []

# Iterate through each <p> tag
for p_tag in p_tags:
    # Find <strong> tag within <p> tag
    strong_tag = p_tag.find('strong')
    if strong_tag and strong_tag.text.strip() == 'When:':
        # Extract the event date text
        event_date = p_tag.text.replace('When:', '').strip()
        event_dates.append(event_date)

# Collect event dates from index 15 to 20 in a list
selected_event_dates = []
for i in range(15, 20):
    if i < len(event_dates):
        selected_event_dates.append(event_dates[i])
    else:
        selected_event_dates.append(f"No event date found at index {i}")

# Print the list of selected event dates
print(selected_event_dates)
```

SCRAPING WEBSITE URLS

```python


# Initialize a list to store URLs
website_urls = []

# Find all <h3> tags
h3_tags = soup.find_all('h3')

# Iterate through each <h3> tag
for h3_tag in h3_tags:
    # Find the <a> tag within <h3>
    a_tag = h3_tag.find('a')
    if a_tag:
        # Extract the URL from the href attribute and append to event_urls
        website_urls.append(a_tag['href'])

# Print all event URLs in a list
print(website_urls[9:14])
```

SCRAPING DESCRIPTIONS

```python
#scraping descriptions
def extract_paragraphs_with_texts(soup, search_texts):
    """Extract paragraphs that contain any of the specified search texts."""
    paragraphs = soup.find_all('p')
    results = []
    for search_text in search_texts:
        for paragraph in paragraphs:
            if search_text in paragraph.get_text():
                cleaned_description = paragraph.get_text(strip=True).replace("Why attend:", "").strip()
                results.append(cleaned_description)
                break  # Stop searching further for this search_text
    return results

# Assuming you have defined your `soup` object using BeautifulSoup

# Search texts (as provided in your example)
search_texts = [
    "If we had to describe Saastr in a few words",
    "The Marketing AI Conference (MAICON) aims to help",
    "IBC bills itself as â€˜the worldâ€™s most inspiring content",
    "DMEXCO (Digital Marketing Exposition & Conference)",
    "This annual conference by Hubspot is a juggernaut"
]

# Extract the specific paragraphs
descriptions = extract_paragraphs_with_texts(soup, search_texts)

# Print results
print("Descriptions list:")
print(descriptions)
```

SCRAPING EVENT TYPES

```python
# Find all <p> tags
p_tags = soup.find_all('p')

# List to store the extracted event types
event_types = []

# Iterate over each <p> tag
for p_tag in p_tags:
    # Get the text content of the <p> tag and strip any leading/trailing whitespace
    p_text = p_tag.get_text().strip()
    # Check if the <p> tag contains 'Event type:'
    if 'Event type:' in p_text:
        # Extract the event type
        event_type = p_text.split('Event type:')[1].strip().split(':')[0]
        # Append the event type to the list
        event_types.append(event_type)

# Print the list of extracted event types
print(event_types[17:22])
```
